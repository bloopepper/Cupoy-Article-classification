```python
from opencc import OpenCC
cc = OpenCC('s2tw')
```

```python
def process_word(dframe, col_name):
    data = dframe[col_name].tolist()
    new_data = []
    for i in data:
        i = str(i).lower()
        new_data.append(cc.convert(i))
    return new_data
```

```python
data_final3['tw_title'] = process_word(data_final3, 'title')
```

```python
data_final3['tw_text'] = process_word(data_final3, 'text')
```

```python
data_final4 = data_final3.drop(['title','text'], axis=1)
data_final4.tail()
```

```python
#處理label
import re
import sys
label_idxs = []
label_vs = []
for index, article in data_final4.iterrows():
    label_str = article['label']
    label_arr = re.split(r'\t+', label_str)
    label_arr = [int(l) for l in label_arr]
    label_idxs.append(index)
    
    v = np.zeros(14).astype('int32')
    for li in label_arr:
        v[li] = 1
    label_vs.append(v)
```

```python
column_list = range(14)

pre_res = ["cat_" + str(sub) for sub in column_list]


res = pd.DataFrame (label_vs,columns=pre_res)
res
```

```python
data_final4 = pd.concat([data_final4, res], axis=1, sort=False)
```
